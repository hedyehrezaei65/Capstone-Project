# Write ground-truth `.txt` files for B6_P_060 line crops

`transcriptions/test_set_line_crops/B6_P_060/` has **36 Kraken line crops** (`B6_P060_L01.jpg` …
`B6_P060_L36.jpg`) but **no ground truth**. This notebook writes one `.txt` per crop, next to the
image (`B6_P060_L01.txt`, …), so the page can join the test set.

**How the crops line up with the page** (verified by eye against each crop):
- `L01–L03` — the heading/date line, which Kraken split into three fragments:
  `South- sea Octobr. ye` / `30` / `1774`
- `L04` — the page/paragraph number `29`
- `L05–L36` — the 32 body lines, in reading order
- The word **`Citato`** in the source text has **no crop of its own** (a margin note Kraken did not
  segment as a line), so it is not written to any file.

The mapping lives in the editable `GROUND_TRUTH` dict below. Lines tagged `# VERIFY` are the
heading/number crops — eyeball them against the images (the preview cell renders each crop with its
assigned text) and tweak the strings if needed before writing.

Runs **locally or on Colab**. On Colab, mount Drive and point `B6_DIR` at the Drive copy.

```python
import os
from glob import glob

# --- Where the B6_P_060 crops live ---
# Local repo checkout:
B6_DIR = "transcriptions/test_set_line_crops/B6_P_060"

# On Colab instead, mount Drive and use the Drive path, e.g.:
# from google.colab import drive; drive.mount('/content/drive')
# B6_DIR = "/content/drive/MyDrive/transcriptions/test_set_line_crops/B6_P_060"

assert os.path.isdir(B6_DIR), f'Not found: {B6_DIR}'
print('Crops dir:', B6_DIR)
print('jpg crops:', len(glob(os.path.join(B6_DIR, '*.jpg'))))
```

## Sanity check — every crop has a mapping, and vice-versa
Confirms the dict keys exactly match the `.jpg` files in the folder before anything is written.

```python
crop_stems = {os.path.splitext(os.path.basename(p))[0]
              for p in glob(os.path.join(B6_DIR, '*.jpg'))}
gt_keys = set(GROUND_TRUTH)

missing_gt   = sorted(crop_stems - gt_keys)   # crops with no text assigned
missing_crop = sorted(gt_keys - crop_stems)   # text assigned to a non-existent crop

print(f'crops on disk : {len(crop_stems)}')
print(f'mapping keys  : {len(gt_keys)}')
print(f'crops missing ground truth : {missing_gt}')
print(f'keys with no crop file     : {missing_crop}')
assert not missing_gt and not missing_crop, 'Fix the mapping so it 1:1 matches the crops'
print('\nOK — mapping is 1:1 with the crops.')
```

## Preview each crop with its assigned ground truth
Scroll through and confirm the text matches the image — especially the `VERIFY` heading crops.

```python
import matplotlib.pyplot as plt
from PIL import Image

stems = sorted(crop_stems)
fig, axes = plt.subplots(len(stems), 1, figsize=(14, 1.5 * len(stems)))
for ax, stem in zip(axes, stems):
    img = Image.open(os.path.join(B6_DIR, stem + '.jpg'))
    ax.imshow(img, cmap='gray')
    ax.set_title(f'{stem}  ->  {GROUND_TRUTH[stem]}', fontsize=9, loc='left')
    ax.axis('off')
plt.tight_layout()
plt.show()
```

## Write the `.txt` files
One UTF-8 `.txt` per crop, same basename, written next to the `.jpg`.

```python
written = 0
for stem, text in GROUND_TRUTH.items():
    txt_path = os.path.join(B6_DIR, stem + '.txt')
    with open(txt_path, 'w', encoding='utf-8') as f:
        f.write(text)
    written += 1

print(f'wrote {written} .txt files into {B6_DIR}')
```

```python
# Verify: every crop now has a non-empty .txt pair
ok = 0
for stem in sorted(crop_stems):
    txt = os.path.join(B6_DIR, stem + '.txt')
    assert os.path.exists(txt), f'missing {txt}'
    with open(txt, encoding='utf-8') as f:
        assert f.read().strip(), f'empty {txt}'
    ok += 1
print(f'{ok} crops now have matching non-empty .txt ground truth')
```
